# Load an information content file from the wordnet_ic corpus.

```python
from nltk.corpus import wordnet as wn
```

```python
from nltk.corpus import wordnet_ic
brown_ic = wordnet_ic.ic('ic-brown.dat')
```

```python
brown_ic
```

## Load Synsets

```python
dog = wn.synset('dog.n.01')
cat = wn.synset('cat.n.01')
```

```python
dog
```

```python
cat
```

## Measure Similarities

### Path (PATH) Similarity

```python
dog.path_similarity(cat, brown_ic)
```

### Leacock-Chodorow (LCH) Similarity

```python
dog.lch_similarity(cat, brown_ic)
```

### Wu-Palmer (WUP) Similarity

```python
dog.wup_similarity(cat, brown_ic)
```

### Resnik (RES) Similarity

```python
dog.res_similarity(cat, brown_ic)
```

### Jiang-Conrath Similarity (JCN) Similarity

```python
dog.jcn_similarity(cat, brown_ic)
```

### Lin (LIN) Similarity

```python
dog.lin_similarity(cat, brown_ic)
```
